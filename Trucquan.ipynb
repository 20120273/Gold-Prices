```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df_gold = pd.read_csv('gold.csv')
df_gold["Date"] = pd.to_datetime(df_gold["Date"])
df_gold.head()
```

```python
#  chnh lch gi vng cao nht vs thp nht trong ngy trong 4 nm gn y
df = pd.DataFrame({
    "date" : df_gold["Date"], 
    "diff" : df_gold["High"] - df_gold["Low"]})

# df.sort_values("diff", ascending=False)
#plot time series
fig, plot = plt.subplots( figsize=(20, 10))
# df1 = df[ (df.date > '2018') & (df.date < '2022') ]

plot.scatter(df["date"], df["diff"])
# plot.scatter(df1["date"], df1["diff"])
plot.grid(True)
plot.set_title("Chnh lch gi vng gia gi vng cao nht v thp nht trong ngy")
plot.set_xlabel("Nm")
plot.set_ylabel("Gi (USD)")
plt.show()
```

```python
# add labels to plot
def value_label(years, diffs):
    for i in range(len(years)):
       plt.text(i, diffs[i], diffs[i], ha = 'center')


df_gold = pd.read_csv('gold.csv')
df = pd.DataFrame({
    "year" : df_gold["Date"].str[:4], 
    "open": df_gold["Open"],
    "close" : df_gold["Close"], 
    "diff": df_gold["Open"] - df_gold["Close"]})

fig, plot01 = plt.subplots(figsize=(18, 9))
plot02 = plot01.twinx()

mean_open_close = df.groupby(by = df.year, as_index=False).mean()

#plot time series
ind = np.arange(len(mean_open_close)) 
width = 0.25

plot01.bar(ind, mean_open_close.open, width, color = 'skyblue', label='Gi m ca')
plot01.bar(ind+width, mean_open_close.close, width, color='grey', label='Gi ng ca')
plot02.plot(mean_open_close['year'], mean_open_close['diff'], color="red")

plot01.set_xlabel('Nm')
plot01.set_ylabel('Gi vng trung bnh')
plot02.set_ylabel(' chnh lch gi vng trung bnh')
plt.xticks(ind + width/2, mean_open_close.year)
plot01.legend(["Gi m ca", "Gi ng ca"])
plt.show()
# value_label(mean_open_close['year'], mean_open_close['diff'])
```
